```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
y
```

```python
#functopnal method
```

```python
plt.plot(x,y)
#plt.show()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
#OO
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.9])

axes1.plot(x,y)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title')

axes2 = fig.add_axes([0.2,0.5,0.3,0.4])
axes2.plot(y,x)
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Title')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
#plt.tight_layout()
axes[0].plot(x,y,'g')
axes[0].set_title('Title')
axes[1].plot(y,x)
axes[1].set_title('Title')
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(4,4))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()

axes=fig.add_axes([0,0,1,1])

axes.plot(x,x**2, label='x squared')
axes.plot(x,x**3, label='x cubed')

axes.legend(loc=(0.4,0.8))
```

```python
fig, axe = plt.subplots()
axe.plot(x,x**2, 'b.-')
axe.plot(x,x**3, 'g--')
```

```python
fig, axes = plt.subplots()
axes.plot(x,x**2,'r.-')
axes.plot(x,x**3,'y--')
```

```python
fig = plt.figure()
axes= fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#A52A2A',lw=1, ls='--', alpha=0.5, marker='o',markersize=20, markerfacecolor='white', markeredgewidth=3, markeredgecolor='yellow')
```

```python
fig = plt.figure()
axes= fig.add_axes([0,0,1,1])
axes.plot(x,y,lw=2,c='g',ls='--')
axes.set_xlim([0,1])
axes.set_ylim([0,2])
```
